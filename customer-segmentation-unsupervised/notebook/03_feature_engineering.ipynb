```python
import os
import pandas as pd

possible_paths = [
    "C:/Users/DELL/Downloads",
    "C:/Users/DELL/Desktop",
    "C:/Users/DELL/Documents"
]

file_path = None

for path in possible_paths:
    try:
        files = os.listdir(path)
        for f in files:
            if "churn" in f.lower() and f.endswith(".csv"):
                file_path = os.path.join(path, f)
                break
    except:
        pass

print("FOUND FILE:", file_path)

df = pd.read_csv("churn_clustered_output.csv")
df.head()
```

```python
df.columns
```

```python
df["Recency"] = df["Customer service calls"]
df["Frequency"] = (
    df["Total day minutes"] +
    df["Total eve minutes"] +
    df["Total night minutes"] +
    df["Total intl minutes"]
)

df["Monetary"] = (
    df["Total day charge"] +
    df["Total eve charge"] +
    df["Total night charge"] +
    df["Total intl charge"]
)

print("RFM Features Added ✔")
```

```python
df["High_Usage"] = (df["Frequency"] > df["Frequency"].median()).astype(int)

df["High_Complaints"] = (df["Customer service calls"] > 3).astype(int)

df["Total_Calls"] = (
    df["Total day calls"] +
    df["Total eve calls"] +
    df["Total night calls"] +
    df["Total intl calls"]
)

df["Avg_Minutes_Per_Call"] = df["Frequency"] / (df["Total_Calls"] + 1)

df["Intl_Usage_Ratio"] = df["Total intl minutes"] / (df["Frequency"] + 1)

print("Behavioral Features Added ✔")
```

```python
df["Charge_Per_Minute"] = df["Monetary"] / (df["Frequency"] + 1)

df["Day_Night_Ratio"] = df["Total day minutes"] / (df["Total night minutes"] + 1)

df["Intl_Call_Ratio"] = df["Total intl calls"] / (df["Total_Calls"] + 1)

print("Derived Ratio Features Added ✔")
```

```python
df.to_csv("processed_feature_engineered_data.csv", index=False)

print("\n🎉 FEATURE ENGINEERING COMPLETED SUCCESSFULLY!")
print("Saved as: processed_feature_engineered_data.csv")
```
