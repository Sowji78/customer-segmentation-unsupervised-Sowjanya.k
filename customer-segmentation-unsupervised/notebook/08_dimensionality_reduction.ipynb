```python
import pandas as pd
df = pd.read_csv("processed_feature_engineered_data.csv")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = df.select_dtypes(include=['int64', 'float64'])

pca_2d = PCA(n_components=2)
pca_result_2d = pca_2d.fit_transform(X)

df['PCA1'] = pca_result_2d[:, 0]
df['PCA2'] = pca_result_2d[:, 1]
plt.figure(figsize=(8,6))
plt.scatter(df['PCA1'], df['PCA2'], s=10, alpha=0.6)
plt.title("PCA - 2D Visualization")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
pca_result_3d = pca_3d.fit_transform(X)

df['PCA3'] = pca_result_3d[:, 2]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("PCA - 3D Visualization")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_result = tsne.fit_transform(X)

df['TSNE1'] = tsne_result[:, 0]
df['TSNE2'] = tsne_result[:, 1]

plt.figure(figsize=(8,6))
plt.scatter(df['TSNE1'], df['TSNE2'], s=10, alpha=0.7)
plt.title("t-SNE Visualization (2D)")
plt.xlabel("TSNE1")
plt.ylabel("TSNE2")
plt.show()
```
