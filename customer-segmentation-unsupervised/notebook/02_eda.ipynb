```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/churn-bigml.csv")

df.head()
```

```python
df.shape        
df.info()        
df.describe()    
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

df[num_cols].hist(figsize=(15,10))
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Count Plot of {col}")
    plt.show()
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Churn', y='Total day minutes', data=df)
plt.show()
```

```python
sns.countplot(x='International plan', hue='Churn', data=df)
plt.show()
```

```python
sns.pairplot(df[['Total day minutes','Total eve minutes','Total night minutes','Churn']], hue='Churn')
plt.show()
```

```python
sns.histplot(df['Total day minutes'], bins=30, kde=True)
plt.show()
```

```python
sns.kdeplot(df['Total day minutes'], shade=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.select_dtypes(include=['int64','float64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
