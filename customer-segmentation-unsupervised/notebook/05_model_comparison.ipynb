```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
df = pd.read_csv("../data/churn-bigml.csv")
df_cluster = df.drop("Churn", axis=1)
df_cluster = pd.get_dummies(df_cluster, drop_first=True)

scaler = StandardScaler()
X = scaler.fit_transform(df_cluster)

kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
df.head()
```

```python
cluster_summary = df.groupby("Cluster").mean(numeric_only=True)
cluster_summary
```

```python
df['Cluster'].value_counts()
```

```python
df.groupby("Cluster")[['Total day minutes','Total eve minutes','Total night minutes']].mean()
df.groupby("Cluster")[['Total day charge','Total eve charge','Total night charge']].mean()
df.groupby("Cluster")['Customer service calls'].mean()
pd.crosstab(df['Cluster'], df['International plan'])
```

```python
pd.crosstab(df['Cluster'], df['State'])
pd.crosstab(df['Cluster'], df['Area code'])
```

```python

cluster_labels = {
    0: "High-Value Premium Customers",
    1: "Budget Occasional Users",
    2: "High-Frequency Low-Spenders",
    3: "At-Risk / Churn-Prone Customers"
}

df['Customer_Type'] = df['Cluster'].map(cluster_labels)
df[['Cluster','Customer_Type']].head()
```
