```python
import pandas as pd

df = pd.read_csv("processed_feature_engineered_data.csv")
cluster_col = "KMeans_Cluster"
cluster_summary = df.groupby(cluster_col).mean(numeric_only=True)
cluster_counts = df[cluster_col].value_counts()

print("===== CLUSTER SUMMARY (MEAN VALUES) =====")
print(cluster_summary)

print("\n===== CLUSTER COUNTS =====")
print(cluster_counts)

print("\n===== INDUSTRY LEVEL CLUSTER INTERPRETATION =====")

for cluster_id in cluster_summary.index:
    
    print(f"\n📌 Cluster {cluster_id}")

    avg_spend = cluster_summary.loc[cluster_id, "Monetary"]
    avg_usage = cluster_summary.loc[cluster_id, "Frequency"]
    complaints = cluster_summary.loc[cluster_id, "Recency"]

    if avg_spend > cluster_summary["Monetary"].mean():
        spend_type = "High Spending"
    else:
        spend_type = "Low / Moderate Spending"

    if avg_usage > cluster_summary["Frequency"].mean():
        usage_type = "High Usage"
    else:
        usage_type = "Low / Moderate Usage"

    if complaints > 3:
        complaint_type = "High Complaints → At Risk"
    else:
        complaint_type = "Low Complaints → Stable"

    if spend_type == "High Spending" and usage_type == "High Usage":
        customer_type = "💎 Premium / High-Value Customers"
    elif complaint_type == "High Complaints → At Risk":
        customer_type = "⚠️ At-Risk / Unhappy Customers"
    elif spend_type == "Low / Moderate Spending" and usage_type == "Low / Moderate Usage":
        customer_type = "🛍️ Budget / Low-Value Customers"
    else:
        customer_type = "🙂 Moderate Users"

    print("Customer Type:", customer_type)
    print("Spending Behavior:", spend_type)
    print("Usage Behavior:", usage_type)
    print("Complaints Behavior:", complaint_type)
    print("-" * 40)
```
