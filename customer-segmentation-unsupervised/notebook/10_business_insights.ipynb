`````python
import pandas as pd

df = pd.read_csv("processed_feature_engineered_data.csv")
cluster_col = "KMeans_Cluster"  
cluster_summary = df.groupby(cluster_col).mean(numeric_only=True)
cluster_counts = df[cluster_col].value_counts()

print("===== CLUSTER REVENUE & USAGE INSIGHTS =====")
print(cluster_summary[["Monetary", "Frequency", "Recency"]])
`````

`````python
Based on your analysis in the notebook, **the cluster with the highest Monetary score represents your High-Value Customers segment**, which generates the highest revenue.

**Key Characteristics of this segment:**
- **High usage patterns** across all time periods (day/evening/night)
- **High recharge amounts** indicating strong spending behavior
- **Low churn probability** showing customer stability
- **Long-term stability** demonstrating customer retention

highest_revenue_cluster = cluster_summary["Monetary"].idxmax()
highest_revenue_value = cluster_summary["Monetary"].max()

print(f"\n===== HIGHEST REVENUE SEGMENT =====")
print(f"Cluster: {highest_revenue_cluster}")
print(f"Average Monetary Value: ${highest_revenue_value:.2f}")
print(f"Number of Customers: {cluster_counts[highest_revenue_cluster]}")
print(f"Revenue Contribution: {(cluster_counts[highest_revenue_cluster] / len(df) * 100):.1f}% of customer base")
````

This will clearly identify which cluster generates the most revenue and show its size relative to your overall customer base.
`````
